```python
import pandas as pd

# Load the dataset
ftu_file_path = 'ref_data/FTU Cell Count Table - Cell_Type_Count.csv'
ftu_data = pd.read_csv(ftu_file_path)

# Filter the data where 'CT ID in CL' starts with 'CL'
ftu_data_clean = ftu_data.dropna(subset=['CT ID in CL'])

ftu_data_clean
```

```python
filtered_ftu_data = ftu_data_clean[ftu_data_clean['CT ID in CL'].str.startswith('CL')]
filtered_ftu_data
```

```python
# Group by 'CT Label in CL' and aggregate 'FTU Label in Uberon'
ftu_aggregated = filtered_ftu_data.groupby('CT Label in CL')['FTU Label in Uberon'].apply(lambda x: '; '.join(x)).reset_index()

# Rename columns as per the user's request
ftu_aggregated.columns = ['Cell type', 'FTUs']
```
